```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
observations = np.array([[1,2], [2,3], [3,4],[5,7],[3,3],
                         [8,7],[3,6],[5,5],[6,8],[7,9],
                         [1,0],[9,5],[4,5],[8,8],[0,1]])
```

```python
n_clusters = 3
```

```python
kmeans = KMeans(n_clusters=n_clusters)
```

```python
kmeans.fit(observations)
```

```python
labels = kmeans.labels_
```

```python
centers = kmeans.cluster_centers_
```

```python
print("Cluster labels for each observation:",labels)
print("Cluster centers:\n", centers)
```

```python
plt.scatter(observations[:,0],observations[:,1], c= labels, cmap='viridis',marker='o')
plt.scatter(centers[:,0],centers[:,1],c='red',marker='x',s=100,label='Centroids')
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
